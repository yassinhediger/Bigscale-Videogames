```
import pandas as pd
import numpy as np
import spacy
import nltk
nltk.download('stopwords')
```

```
df = pd.read_csv('https://raw.githubusercontent.com/yassinhediger/Bigscale-Videogames/master/data/rawdata/videogamesales.csv')
df.head()
```

# Column renaming

```
df = df.rename(columns={"Rank": "game_global_ranking", "Name": "game_title", "Platform": "console","Genre": "genre", "Year": "year_of_release", "Publisher": "publisher", "NA_Sales": "northamerican_sales", "EU_Sales": "europe_sales", "JP_Sales": "japan_sales", "Other_Sales":"othercontry_sales", "Global_Sales;": "world_sales"})
df.head()
```

# Text cleaning


```
#Lower all the letters from Name column
df["game_title"] = df["game_title"].str.lower()
df["console"] = df["console"].str.lower()
df["genre"] = df["genre"].str.lower()
df["publisher"] = df["publisher"].str.lower()
#Removing White space
df["game_title"]= df["game_title"].str.replace(" ", "") 
#Removing special character
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","‚Äì"]
for char in spec_chars:
    df['game_title'] = df['game_title'].str.replace(char, ' ')

df.head()





df.head(100)
```
